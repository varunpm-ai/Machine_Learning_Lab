```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

np.random.seed(42)

X = np.random.rand(100, 1)

y = (X[:, 0] > 0.5).astype(int)

k_values = [1, 2, 3, 5, 20, 30]

plt.figure(figsize=(15, 10))

for i, k in enumerate(k_values, 1):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X, y)
    
    y_pred = model.predict(X)
    
    plt.subplot(3, 2, i)
    
    plt.scatter(X, np.zeros_like(X), c=y, cmap='coolwarm', edgecolors='k', label='Original')
    
    plt.scatter(X, np.ones_like(X)*0.1, c=y_pred, cmap='coolwarm', marker='x', label='Predicted')
    
    plt.title(f"k = {k}")
    plt.xlabel("X")
    plt.yticks([])
    plt.grid()

plt.tight_layout()
plt.show()
```
